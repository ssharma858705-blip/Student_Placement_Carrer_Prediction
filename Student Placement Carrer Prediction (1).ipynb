```python
print("Average CGPA:", np.mean(df["cgpa"]))

print("Median CGPA:", np.median(df["cgpa"]))

print("Max CGPA:", np.max(df["cgpa"]))

print("Min CGPA:", np.min(df["cgpa"]))

print("Standard Deviation:", np.std(df["cgpa"]))
```

```python
print(df["placement_status"].value_counts())
```

```python
branch_placement = df.groupby("branch")["placement_status"].value_counts()

print(branch_placement)
```

```python
high_cgpa = df[df["cgpa"] > 8]

print(high_cgpa.head())
```

```python
corr = df.corr(numeric_only=True)

print(corr)
```

```python
branch_count = pd.pivot_table(
    df,
    index="branch",
    values="student_id",
    aggfunc="count"
)

print(branch_count)
```

```python
sns.histplot(df["cgpa"], kde=True)

plt.title("CGPA Distribution")

plt.show()
```

```python
sns.countplot(x="branch", hue="placement_status", data=df)

plt.title("Branch vs Placement")

plt.show()
```

```python
sns.boxplot(x="placement_status", y="coding_skills_score", data=df)

plt.title("Coding Skills vs Placement")

plt.show()
```

```python
top_students = df.groupby("student_id")["cgpa"].max().sort_values(ascending=False).head(10)

colors = ["red","yellow","blue","purple","green","orange","pink","cyan","brown","grey"]

plt.bar(top_students.index.astype(str), top_students.values, color=colors)

plt.xticks(rotation=45, ha="right")

plt.title("Top 10 Students by CGPA")

plt.xlabel("Student ID")
plt.ylabel("CGPA")

plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True)

plt.show()
```

```python
df["placement_status"].value_counts().plot.pie(autopct="%1.1f%%")

plt.title("Placement Distribution")
plt.ylabel("")

plt.show()
```
